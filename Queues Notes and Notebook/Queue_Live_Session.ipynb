# <font color='blue'> Queues </font>

A queue is an ordered collection of items where the insertion of new items happens at one end. This end is generally known as <rear>. The deletion of any existing items can happen at only one end, this is generally called the front. 
    
Once the item inserted inside the queue, it starts at the rear and makes its way toward the front, waiting until that time when it is the next item to be removed from the queue.

In queue, the most recently inserted item in the queue should be available at the end of the collection. 
    
The item that has been in the collection the longest should be available at the front. This ordering principle is also called as FIFO, first-in first-out. 
    
## <font color='blue'> Queue: Abstract Data Types </font>
    
The queue operations are given below.

* Queue() creates a new queue that is empty. It needs no parameters and returns an empty queue.

* enqueue(item) adds a new item to the rear of the queue. It needs the item and returns nothing.

* dequeue() removes the front item from the queue. It needs no parameters and returns the item. The queue is modified.

* is_empty() tests to see whether the queue is empty. It needs no parameters and returns a boolean value.

* size() returns the number of items in the queue. It needs no parameters and returns an integer.

        While (!isempty(Stack1))

        returnval = Pop from Stack2

        return returnval

### <font color='blue'> Stack Class </font>

```python
class Stack:

    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        return self.items.pop()

    def size(self):
        return len(self.items)

    def is_empty(self):
        return self.items == []
```

```python
class Queue(object):

    def __init__(self):

        # Initializing two Stacks
        self.stack1 = Stack()
        self.stack2 = Stack()

    def enqueue(self, item):
        self.stack1.push(item)

    def dequeue(self):
        
        if self.stack1.size() == 0 and self.stack2.size() == 0:
            print("Queue is Empty")
            return
 
        # if s2 is empty and s1 has elements
        elif self.stack2.size() == 0 and self.stack1.size() > 0:
            while self.stack1.size():
                temp = self.stack1.pop()
                self.stack2.push(temp)
            return self.stack2.pop()
 
        else:
            return self.stack2.pop() 
```

# <font color='blue'> Priority Queue </font>

In this section, we will be understanding one of the most broadly useful sophisticated data structures, the priority queue. 

A priority queue is designed for applications in which elements have a priority value, or key, and each time we need to select an element from S, we want to take the one with highest priority.

A priority queue is a data structure that maintains a set of elements S, where each element v â S has an associated value key(v) that denotes the priority of element v; smaller keys represent higher priorities. Priority queues support the addition and deletion of elements from the set, and also the selection of the element with smallest key

Priority queues are useful in multiple situations such as priority based task scheduler, graph algorithms etc. 

For our purposes here, it is a useful illustration of the analysis of a data structure that, unlike lists and arrays, must perform some nontrivial processing each time it is invoked.


## <font color='blue'> Implementation of Priority Queue </font>


```python
class Node:
    def __init__(self, data, priority):
        self.data = data
        self.priority = priority
        self.next = None

class Queue:
    
    def __init__(self):
        self.front = None
        self.rear = None

    # This method will insert data in the queue. 
    def enqueue(self, val, prior):
        # Creating the node for insertion
        tmp = Node(val, prior)
        # Checking if the Queue is empty
        if self.rear is None:
            self.front = self.rear = tmp
            return
        
        # if queue has only one element 
        # Inserting the element and rearraning the front
        if (self.front.priority > prior):
            tmp.next = self.front
            self.front = tmp
            
        # Finding the right position to insert the elelemnt in the queue
        # this will ensure the priority of elelemnts are arranged.  
        else:
            p = self.front
            while p.next != None and p.next.priority <prior:
                p = p.next
            
            tmp.next = p.next
            p.next = tmp

    def isempty(self):
        # 1. If front is None
        #a. Queue is empty 
        if self.front is None:
            return True
        #2. Queue has some element. 
        else:
            return False


    def dequeue(self):
        #1. Check if queue is empty
        if self.isempty():
            print("Queue is empty")
            self.front = None
            self.rear = None
        #2. If the queue has some element then deletion can happen
        # we need to update the front as well 
        else:
            #a. save the front reference in another variable
            #b. Move the front pointer to the next node
            #c. Print the data
            #d. delete the node data
            tmp = self.front
            print("Deleted element is element of Queue is :", end =" ")
            print(tmp.data)
            self.front = tmp.next
            del tmp
            if self.isempty():
                self.rear = None

q = Queue()

q.enqueue("P1", 10)
q.enqueue("P2", 12)
q.enqueue("P3", 1)
q.enqueue("P4", 2)

q.dequeue()
q.dequeue()
q.dequeue()
q.dequeue()
q.dequeue()
```
